```python
directory = '../data/eleanor/s0007/'

filenames = os.listdir(directory)

# Dictionary to store matching filenames for each ID
id_to_files = {}

# Iterate over each ID in the DataFrame
for id_value in tqdm(cdips_lcs['tic_id']):
    pattern = re.compile(rf'{id_value}')
    matching_files = [filename for filename in filenames if re.search(pattern, filename)]
    id_to_files[id_value] = matching_files
```

```python
files2 = glob.glob('../data/eleanor/cdips-lcs/*.fits')
np.random.shuffle(files2)

fo = '../models/cdips/'
os.makedirs(fo,exist_ok=True)
t0 = []
t2 = []
tic = []
for i in tqdm(files2[0:5000]):
    lc, info = import_lightcurve(i)
    t = lc['TIME']
    t0.append(t[0] + 5)
    t2.append(t[-1] - 5)
    tic.append(info['TIC_ID'])
    f = lc['PCA_FLUX']/np.nanmedian(lc['PCA_FLUX'])
    fe = lc['FLUX_ERR']
    np.save(f"{fo}/{info['TIC_ID']}_sector07.npy", np.array([t,f,fe]))
```

### Creating a CDIPS stellar variability catalog

- Then a second one from `t[-1] - 5`

```python
data = pd.DataFrame(data=[tic,t0]).T
data.columns = ['TIC','tpeak']
data.TIC = data.TIC.astype(int)
```
